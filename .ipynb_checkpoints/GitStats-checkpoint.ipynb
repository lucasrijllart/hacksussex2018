```python
from git import Repo

repo = Repo("~/hackathon/Sleep-Wake")

commit_list = repo.iter_commits("master")

print(dir(next(commit_list)))

authors = []
messages = []
for i in commit_list:
#     print(dir(i))
    authors.append(i.author)
    messages.append(i.message)
    
    
authors = list(set(authors))
print(authors)

print(messages)

print(dir(authors[0]))
```

```python
repo = Repo("~/hackathon/Sleep-Wake")

commit_list = repo.iter_commits("master")

author_commit_count = {}
def add_authors(dictionary):
    for i in authors:
        dictionary[i.name] = 0
        
add_authors(author_commit_count)

def commit_count_filtered(dictionary, date, interval):
    start_date = date
    end_date = date + timedelta(days=interval)
    for i in commit_list:
        if start_date<i.authored_datetime and i.authored_datetime<end_date:
            if i.author.name in dictionary:
                dictionary[i.author.name] += 1

commit_count_filtered(author_commit_count, date=datetime(2017, 6, 30, tzinfo=timezone.utc), interval=7)

print(author_commit_count)    
```
